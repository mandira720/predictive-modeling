```python
m=data['depth'].median()
data['depth']=data['depth'].fillna(m)
```

```python
data=data.drop(columns=['Unnamed: 0'])
```

```python
for feature in data.columns:
    if data[feature].dtype =='float64'or data[feature].dtype =='int64':
        print('feature:',feature)
        print(pd.Categorical(data[feature].unique()))
```

```python
for feature in data.columns:
    if data[feature].dtype =='object':
        print('feature:',feature)
        print(pd.Categorical(data[feature].unique()))
        
data=pd.get_dummies(data,columns=["clarity"])
data['cut']=data['cut'].replace({'Fair': 1,'Ideal' :2,'Good' :3,'Very Good' :4,'Premium' :5})
data['color']=data['color'].replace({'J': 1,'I' :2,'H' :3,'G' :4,'F' :5,'E' :6,'D' :7})
```

```python
for feature in data.columns:
    if data[feature].dtype =='uint8':
        data[feature]=data[feature].astype('int64')
```

```python
q1=data.quantile(0.25)
q3=data.quantile(0.75)
iqr=q3-q1
((data<(q1-1.5*iqr))|(data>(q3+1.5*iqr))).sum()
```

```python
df=data.copy()
```

```python
x=df.drop('price',axis=1)
y=df.pop('price')
xy=x.copy()
```

```python
scaler=MinMaxScaler()
scaler.fit(x)
x=scaler.transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
linear=LinearRegression()
linear.fit(x_train,y_train)
```

```python
scores=cross_val_score(linear,x_test,y_test,scoring='r2',cv=10)
scores.mean()
```

```python
scores=cross_val_score(linear,x_train,y_train,scoring='r2',cv=10)
scores.mean()
```

```python
linear.score(x_train,y_train)
```

```python
linear.score(x_test,y_test)
```

```python
ytrain_predict=linear.predict(x_train)
ytrain_predict
ytest_predict=linear.predict(x_test)
```

```python
y_train
```

```python
d = y_train - ytrain_predict
mse_f = np.mean(d**2)
mae_f = np.mean(abs(d))
rmse_f = np.sqrt(mse_f)
r2_f = 1-(sum(d**2)/sum((y_train-np.mean(y))**2))

print("Results by manual calculation:")
print("MAE:",mae_f)
print("MSE:", mse_f)
print("RMSE:", rmse_f)
print("R-Squared:", r2_f)
```

```python
d = y_test - ytest_predict
mse_f = np.mean(d**2)
mae_f = np.mean(abs(d))
rmse_f = np.sqrt(mse_f)
r2_f = 1-(sum(d**2)/sum((y_train-np.mean(y))**2))

print("Results by manual calculation:")
print("MAE:",mae_f)
print("MSE:", mse_f)
print("RMSE:", rmse_f)
print("R-Squared:", r2_f)
```
